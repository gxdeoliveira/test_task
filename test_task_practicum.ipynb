### Task 1. Working with data
This task is divided into eleven (11) steps. Each one will be responded and pointed bellow.

```python
# importing the libraries for the task
import pandas as pd
import numpy as np
import datetime
```

```python
# 1.1. Download the data set movie_metadata.csv, which contains data about films from IMDb (Internet Movie Database).
movie_metadata = pd.read_csv("movie_metadata.csv")
```

```python
movie_metadata.head()
```

```python
# 1.2. The duration column contains data on the film length. How many missing values are there in this column?
movie_metadata["duration"].isna().sum()
```

```python
# 1.3. Replace the missing values in the duration column with the median value for this column.
movie_metadata["duration"] = movie_metadata["duration"].fillna(movie_metadata["duration"].median())
```

```python
movie_metadata["duration"].isna().sum()
```

```python
#1.4. What is the average film length? Give the answer as a floating-point figure rounded to two decimal places.
length_movies = movie_metadata["duration"].mean()
str(round(length_movies, 2))
```

```python
# 1.5. Create a movie_duration_category column, which will contain three categories depending on the film length:
# • Category "1. <90" if the film is less than 90 minutes long
# • Category "2. 90–120" if the film is between 90 minutes and two hours long (inclusively)
# • Category "3. >120" if the film is more than two hours long

conditions = [
    (movie_metadata["duration"] < 90),
    (movie_metadata["duration"] >= 90) & (movie_metadata['duration'] <= 120),
    (movie_metadata['duration'] > 120)
    ]
values = ['<90', '90-120', '>120']
movie_metadata["movie_duration_category"] = np.select(conditions, values)
movie_metadata
```

```python
# Filling the missing data in "title_year"
movie_metadata["title_year"] = movie_metadata["title_year"].fillna(movie_metadata["title_year"].median().astype(int))
movie_metadata["title_year"] = movie_metadata["title_year"].astype(int)
```

```python
movie_metadata['title_year']
```

```python
# 1.7. How many films between 90 minutes and two hours long were released in 2008?
pd.crosstab(movie_metadata['movie_duration_category'], movie_metadata['title_year']==2008)
```

```python
#1.11. The budget column contains the film's budget. What is the median budget for all the films listed? Give the answer as an integer.
movie_metadata["budget"] = movie_metadata["budget"].str.replace('[\$\, \.]', '', regex=True)
movie_metadata["budget"].median()
```

### Task 2. Problem-solving
This task is divided into nine (9) steps. Each one will be responded and pointed bellow.

```python
# 1. Download the event_data.csv dataset
event_data = pd.read_csv("event_data.csv")
```

```python
event_data.head()
```

```python
event_data.info()
```

```python
# 2. Highlight user cohorts based on the week of registration in the application. The cohort identifier should 
# be the week ordinal (for example, the week from July 29 to August 4 should have identifier 31).

# turn the event_date to python date time object
event_data['event_date'] = pd.to_datetime(event_data['event_date'],format='%Y-%m-%d %H:%M:%S')

# creating the user_cohort by week
event_data['week'] = [int(i) + 1 for i in event_data['event_date'].dt.strftime('%W')]
event_data
```

```python
#3. How many unique users in the cohort with ID 33?

#1st. set the event_type as registration
registration_event = event_data[event_data['event_type'] == 'registration']
#2nd. extract user_id/week to relate user_id and the week
registration_event = registration_event[['user_id', 'week']]
registration_event.columns = ['user_id', 'registration_week']
registration_event
```

```python
# merge the week of registration for each user_id in the df
event_data = event_data.merge(registration_event, how='left', on='user_id')
#checking the cohort lifetime
event_data['cohort_lifetime'] = event_data['week'] - event_data['registration_week']
```

```python
event_data
```

```python
# calculate the number of unique user_id for every registration week 

cohorts = event_data.pivot_table(index = ['registration_week','cohort_lifetime'], 
                         aggfunc = {'user_id':'nunique'}).reset_index()
cohorts
```

```python
initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0]
initial_users_count = initial_users_count[['registration_week','user_id']]
initial_users_count = initial_users_count.rename(columns={'user_id':'cohort_users'})
initial_users_count 
```

```python
# 4. For each event, highlight the indicator lifetime - the weekly lifetime of the cohort. The lifetime indicator 
# is calculated based on the serial number of the week in which the event is committed, relative to the week of 
# registration. For example, an event committed on August 3 by a user from a cohort of registrants at 31 weeks 
# will be committed on the zero week of lifetime, and an event committed by the same user on August 5 will be 
# committed on the first week of lifetime).
event_data.pivot_table(index =['cohort_lifetime','registration_week'], 
                       aggfunc = {'user_id':'nunique'}).reset_index()
```

```python
# 5. Build a summary table of changes in the Retention Rate for cohorts depending on lifetime.

cohorts = cohorts.merge(initial_users_count,on='registration_week')
cohorts['retention'] = cohorts['user_id']/cohorts['cohort_users'] 
retention_pivot = cohorts.pivot_table(index='registration_week',
                                      columns='cohort_lifetime',
                                      values='retention',
                                      aggfunc='sum') 
```
